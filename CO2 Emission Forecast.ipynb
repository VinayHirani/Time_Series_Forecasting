```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Read in the Dataframe

```python
df = pd.read_csv("MER_T12_06.csv")
df.head()
```

#### Since the YYYYMM column is a string i have formatted it to the correct format so i can parse the dates properly into the dataframe

```python
dateparse = lambda x: pd.to_datetime(x, format='%Y%m', errors = 'coerce')
df = pd.read_csv("MER_T12_06.csv", parse_dates=['YYYYMM'], index_col='YYYYMM', date_parser=dateparse) 
df.head()
```

#### Removing any Null Datetime index Row 

```python

ts = df[pd.Series(pd.to_datetime(df.index, errors='coerce')).notnull().values]
ts.head(15)
```

```python
ts['Value'] = pd.to_numeric(ts['Value'] , errors='coerce')
ts.head()
```

```python
ts.dropna(inplace = True)
```

```python
Energy_sources = ts.groupby('Description') #Grouping by so i can plot each type of energy source
Energy_sources.head()
```

#### Plotting 

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20, 25
fig, ax = plt.subplots()
for desc, group in Energy_sources:
    group.plot(y='Value', label=desc,ax = ax, title='Carbon Emissions per Energy Source', fontsize = 20)
    ax.set_xlabel('Time(Monthly)')
    ax.set_ylabel('Carbon Emissions in MMT')
    ax.xaxis.label.set_size(20)
    ax.yaxis.label.set_size(20)
    ax.legend(fontsize = 16)
```

### There is a clear seasonality and trend here. Let's try holt winters methods of Simple,Double,Triple Exponential Smoothing

```python
# Simple Exp Smoothing 
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

```

```python
span = 12
alpha=2/(span+1)
df_total_emission['SES12'] = SimpleExpSmoothing(df_total_emission['Value']).fit(smoothing_level=alpha,
                                                                               optimized=False).fittedvalues.shift(-1)
df_total_emission.head()
```

```python
#Plotting SES12 and Value 
df_total_emission['Value'].plot(figsize=(12,5),legend=True)
df_total_emission['SES12'].plot(legend=True,title='SES12(HW)')
```

```python
# Double Exponential Smoothing 
from statsmodels.tsa.holtwinters import ExponentialSmoothing
df_total_emission['DES12ADD'] = ExponentialSmoothing(df_total_emission['Value'],
                                                     trend='add').fit().fittedvalues.shift(-1)
df_total_emission.head()
```

```python
# Plotting DES ,SES , Values 
df_total_emission['Value'].plot(figsize=(12,5),legend=True)
df_total_emission['SES12'].plot(legend=True,title='SES12(HW)')
df_total_emission['DES12ADD'].plot(legend=True,title='DES12')
```

```python
df_total_emission[['Value','SES12','DES12ADD']].iloc[480:].plot(figsize=(12,5))
```

```python
df_total_emission['TES12ADD'] = ExponentialSmoothing(df_total_emission['Value'],
                                                    trend='add',seasonal='add',
                                                    seasonal_periods=12).fit().fittedvalues
df_total_emission.head()
```

```python
df_total_emission[['Value','SES12','DES12ADD','TES12ADD']].plot(figsize=(12,5))
```

```python
df_total_emission[['Value','SES12','DES12ADD','TES12ADD']].iloc[480:].plot(figsize=(12,5))
```

```python
train = df_total_emission.iloc[:492]
test = df_total_emission.iloc[492:]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(train['Value'],order=(1,1,1),
               seasonal_order=(1,0,1,12),enforce_invertibility=False)
results = model.fit()
results.summary()
```

```python
start = len(train)
end = len(train) + len(test) - 1 
```

```python
predictions = results.predict(start=start,end=end,
                             dynamic=False,typ='levels').rename('SARIMA Predictions')
predictions 
```

```python
test['Value'].plot(figsize=(12,5),legend=True)
predictions.plot(legend=True)
```

```python
from statsmodels.tools.eval_measures import rmse

error = rmse(test['Value'],predictions)
error 
```

### Forecast into the future 

```python
model = SARIMAX(df_total_emission['Value'],order=(1,1,1),
                  seasonal_order=(1,0,1,12),enforce_invertibility=False)
results = model.fit()
results.summary()
```

```python
predictions = results.predict(start=len(df_total_emission['Value']), 
                             end = len(df_total_emission['Value'])+11,typ='levels').rename(" Forecasted")
predictions 
```

```python
title="Monthly CO2 Emissions Forecast"
xlabel = 'Year'
ylabel = "Million Metric Tons of CO2"

ax = df_total_emission['Value'].plot(figsize=(12,5),legend=True,title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```
